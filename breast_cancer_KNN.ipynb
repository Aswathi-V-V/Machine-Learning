# **BREAST CANCER PREDICTIONbcancer.webp**

The dataset above is a breast cancer diagnostic dataset containing measurements from breast cancer tumors. It includes various features that describe the characteristics of cell nuclei present in breast cancer biopsies. Each row represents a different patient, and columns include:

* ID: Patient identifier.
* Diagnosis: Whether the tumor is malignant (M) or benign (B).
* Mean Features: Includes measurements like radius, texture, perimeter, area, smoothness, compactness, concavity, and concave points.
* Worst Features: Similar measurements as mean features but calculated for the worst (largest) values for each patient, including radius, texture, perimeter, area, smoothness, compactness, concavity, and concave points.

This dataset is used for training models to predict the malignancy of breast cancer based on these features.

**IMPORTING LIBRARIES**

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
```

**LOAD DATASET**

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/breast-cancer.csv')
df
```

**DATA PREPROCESSING**

```python
#Checking the missing values
df.isna().sum()
```

```python
#Check the datatypes
df.dtypes
```

```python
df['diagnosis']=df['diagnosis'].str.replace('M','1')
df['diagnosis']=df['diagnosis'].str.replace('B','0')
df
```

```python
df['diagnosis']=df['diagnosis'].astype(int)
```

```python
#split into X and y
X=df.iloc[:,2:].values
X
```

```python
y=df.iloc[:,1]
y
```

**SCALING**

```python
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

**TRAIN TEST SPLIT**

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

**MODEL CREATION**

```python
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
y_test.values
```

```python
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
